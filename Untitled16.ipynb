```python
stocks = pd.DataFrame({"Date": tcs_data["Date"] ,"TCS": tcs_data["Close"],
                      "INFY": infy_data["Close"],
                      "NIFTY_IT": nifty_it["Close"]})
```

```python
"""
Checking Stationarity of each series

p-value > 0.05: Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.
p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.

"""

from statsmodels.tsa.stattools import adfuller
def Stationarity_test (Series) :
    X = stocks[Series]
    result = adfuller(X,autolag='AIC')
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('#Lags Used : %f' % result [2])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
    
    

Series = raw_input("Enter the series you want to test for stationarity:\nTCS \nINFY \nNIFTY_IT :" )    
    
Stationarity_test(Series)        
```

```python
"""
Creating Moving Average as per user input (e.g. 4,16,24,32..52)
Plotting the MA created using matplotlib for individual series : TCS , INFY, NIFTY_IT
"""

def Moving_Avg(week,choice):
    days = week*7 
    #MA = pd.rolling_mean(tcs_data['Close'],days).plot(x='Date')
    #days = week*7
    #MA = pd.rolling_mean(stocks,days)
    #return MA
    if choice == 'Plot' :
        Series_inp = raw_input("Enter the series you want to plot: \n TCS \n INFY \n NIFTY_IN  \n")
        pd.rolling_mean(stocks[Series_inp],days).plot()
        stocks[Series_inp].plot()
        
    elif choice == 'Data' :
        
        MA = pd.rolling_mean(stocks,days)
        return MA
    elif choice == 'Both' :
        Series_inp = raw_input("Enter the series you want to plot: \n TCS \n INFY \n NIFTY_IN  \n")
        pd.rolling_mean(stocks[Series_inp],days).plot()
        stocks[Series_inp].plot()
        MA=pd.rolling_mean(stocks,days)
        return MA

    

choice = raw_input("Choose from option: \n1.Data \n2.Plot \n3.Both: \n ")
week_input = int(raw_input("Enter no. of week to see Moving Average: "))
#Series_inp = raw_input("Enter the series you want to plot: ")

#Moving_Avg(pd.to_numeric(week_input))     
Moving_Avg(week_input,choice)
   
#tcs_data['Close'].plot()
```
